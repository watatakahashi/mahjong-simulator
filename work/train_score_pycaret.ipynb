```python
import numpy as np
import pandas as pd
from pycaret.classification import *
```

```python
df = pd.read_csv('score.csv', header=None)
df = df.iloc[:, 9:14]
df = df.rename(columns={9:'AB', 10:'AC', 11:'AD', 12:'score', 13:'round'})
```

```python
df
```

```python
df = df[df['round'] == 3]
df = df.reindex(np.random.permutation(df.index)).reset_index(drop=True)
# 分割
ratio = 0.8
p = int(ratio * len(df)) # 分割点を計算
train_df = df.iloc[:p, :]
test_df = df.iloc[p:, :]
```

```python
exp_clf101 = setup(data = train_df, target = 'score', session_id=123, silent=True) 
```
